```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies_df = pd.read_csv('resources/movies.csv')
movies_df
```

```python
oscar_details_df = pd.read_csv('resources/oscar_details.csv')
oscar_details_df
```

```python
movie_ratings_df = pd.read_csv('resources/movie_ratings.csv')
movie_ratings_df
```

```python
movie_revenue_df = pd.read_csv('resources/movie_revenue.csv')
movie_revenue_df
```

```python
# Merging all the DataFrames together which is needed to perform the analysis
Oscar_combined_df = pd.merge(movies_df,movie_ratings_df,on="movie_id")
Oscar_combined_df = pd.merge(Oscar_combined_df,movie_revenue_df,on="movie_id",how='outer')
Oscar_combined_df = pd.merge(Oscar_combined_df,oscar_details_df,on="movie_id")

```

```python
Oscar_combined_df
```

```python
# Split the Genre list by comma and create a list of unique generes 
# It is done since we will be needing to write queries to group by each genre 

genre_list = sorted(Oscar_combined_df['genre'].str.split(',',expand=True).stack().str.strip().unique().tolist())
genre_list
```

```python
# Create a column for each Genre and fill with True/ False values

for genre in genre_list:
    Oscar_combined_df[f'genre_{genre}'] = Oscar_combined_df['genre'].str.contains(genre)

# Drop the original genre column since we have already split it up
Oscar_combined_df.drop('genre',axis=1,inplace=True)
Oscar_combined_df
```

```python
# Extract only the main producer from the producer list and create a seperate column for it
Oscar_combined_df['movie_producer_main'] = Oscar_combined_df['movie_producer'].str.split(',').str[0]
Oscar_combined_df['movie_producer_main']
```

```python
# Extract only the main director from the director list and create a seperate column for it
Oscar_combined_df['movie_director_main'] = Oscar_combined_df['movie_director'].str.split(',').str[0]
Oscar_combined_df['movie_director_main']
```

```python
# Extract the first 2 actors in the list and create 2 new columns for them
Oscar_combined_df['movie_actor_1'] = Oscar_combined_df['movie_actor'].str.split(',',expand=True)[0]
Oscar_combined_df['movie_actor_2'] = Oscar_combined_df['movie_actor'].str.split(',',expand=True)[1]
Oscar_combined_df['movie_actor_1']
Oscar_combined_df['movie_actor_2']
```

```python
# Split the award column into 2 new columns which helps to analyse the data based on winner/non-winner category
Oscar_combined_df['award_Nominee'] = Oscar_combined_df['award'].str.contains('Nominee').astype(int)
Oscar_combined_df['award_Winner'] = Oscar_combined_df['award'].str.contains('Winner').astype(int)
# Drop the original 'award' column
#Oscar_combined_df.drop(columns=['award'], inplace=True)
```

```python
# Calculate the average movie Ratings :
# Rescale the ratings to a common scale( out of 10)

Oscar_combined_df['imdb_rating_scaled'] = Oscar_combined_df['imdb_rating']
Oscar_combined_df['tomatometer_rating_scaled'] = Oscar_combined_df['tomatometer_rating'] / 10
Oscar_combined_df['audience_rating_scaled'] = Oscar_combined_df['audience_rating'] / 10

def calculate_average_rating(row):
    if row['tomatometer_rating'] != 0:
        # If tomatometer_rating is not 0, calculate the average of imdb_rating, tomatometer_rating, and audience_rating
        return (row['imdb_rating_scaled'] + row['tomatometer_rating_scaled'] + row['tomatometer_rating_scaled']) / 3
    else:
        # If tomatometer_rating is 0, calculate the average of imdb_rating only
        return (row['imdb_rating'] ) 

# Apply the function to each row of the DataFrame to create a new column 'avg_rating'
Oscar_combined_df['avg_rating'] = Oscar_combined_df.apply(calculate_average_rating, axis=1)
Oscar_combined_df
```

```python
# Question 2: Are there any noticable directors who have a higher likelihood of winning an Oscar?

# Filter out rows where movie_director_main is blank
cleaned_directors_df = Oscar_combined_df[Oscar_combined_df['movie_director_main'] != ' ']
cleaned_directors_df.reset_index(drop=True,inplace=True)

# Group by director and calulate the number of nominations and wins
director_oscar_count = cleaned_directors_df.groupby('movie_director_main')[['award_Nominee', 'award_Winner']].sum()


# calculate the total number of nominations and wins for each director
director_oscar_count['total_director_count'] = director_oscar_count['award_Nominee'] + director_oscar_count['award_Winner']

top_directors = director_oscar_count.sort_values(by='total_director_count', ascending=False).head(10)

top_directors
```

```python
# Plot for Question 2: Are there any noticable directors who have a higher likelihood of winning an Oscar?

sns.set_style("whitegrid")

# Create plot
plt.figure(figsize=(10,6))
sns.histplot(data= cleaned_directors_df[cleaned_directors_df['movie_director_main'].isin(top_directors.index)],x='movie_director_main', hue='award',palette=['#4e79a7','#f28e2c'], alpha=0.7 )


# Set plot title and axis labels
plt.title('Distribution of Oscar Nominations and Wins by Director')
plt.xlabel('Directors')
plt.ylabel('Total Oscar Nominations and Wins')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Question : Are movies with higher production budget more likely to win Oscars
# Create a scatter plot of production budget vs. number of Oscars won
from bokeh.plotting import figure, output_file, show, save
from bokeh.models import ColumnDataSource, HoverTool

# Filter the dataframe to include only movies with production_budget > 0 and their respective winners
filtered_df = Oscar_combined_df.loc[Oscar_combined_df['award_Winner'].isin([0, 1])]
filtered_df = filtered_df.loc[filtered_df['production_budget'] > 0]

# Scale down the production_budget values by dividing by a factor of 1,000,000
filtered_df['production_budget'] = filtered_df['production_budget'] / 1000000

# Create a ColumnDataSource from the filtered dataframe
source = ColumnDataSource(filtered_df)

# set the output file name and format
output_file('production_budget_vs_oscar_winner.html', title='Production Budget vs. Oscar Winner')


# Create a figure object and add the scatter plot
p = figure(title='Production Budget vs. Oscar Winner', 
           x_axis_label='Production Budget (in millions)',
           y_axis_label='Oscar Winner (0 = No, 1 = Yes)')
p.scatter(x='production_budget', y='award_Winner', 
          size=8, alpha=0.5, source=source,
          legend_label='award_winner')

# Add a HoverTool to display the movie title as a tooltip
hover = HoverTool(tooltips=[('Movie', '@movie_name')])
p.add_tools(hover)


# Set the legend location and orientation
p.legend.location = 'top_left'
p.legend.orientation = 'vertical'

# save the plot to file
save(p)

# Show the plot
show(p)
```

```python
# Question : Are movies with good average ratings more likely to win oscars
Oscar_combined_df['avg_rating'].isna().sum()

filtered_df = Oscar_combined_df.sort_values(by='avg_rating', ascending=False)
Oscar_data = filtered_df[filtered_df['award_Winner']==1]

# Calculate the percentage of movies with good average ratings that have won an Oscar
Percent_good_oscarratings = (Oscar_data['avg_rating'] >= filtered_df['avg_rating'].mean()).sum() / len(filtered_df) * 100

print(f"{Percent_good_oscarratings:.2f}% of movies with good average ratings have won an Oscar")
```

```python
# Question : Percentage of Movies with good average ratings that won the oscars Vs that did not win
# Define a function to determine if a movie has a good average rating
def is_good_rating(row):
    return row['avg_rating'] >= 7

# Create a new column indicating if a movie has a good rating or not
filtered_df['good_rating'] = filtered_df.apply(is_good_rating, axis=1)

# Group the movies by whether or not they have a good rating and whether or not they won an Oscar
grouped = filtered_df.groupby(['good_rating', 'award_Winner'])['movie_id'].count().reset_index()

# Calculate the percentage of movies with good ratings that have won an Oscar
total_good_movies = grouped[grouped['good_rating'] == True]['movie_id'].sum()
good_movies_won_oscar = grouped[(grouped['good_rating'] == True) & (grouped['award_Winner'] == 1)]['movie_id'].sum()
percentage = good_movies_won_oscar / total_good_movies * 100
# Create a bar plot of the results
labels = ['Good Ratings', 'Bad Ratings']
values = [percentage, 100 - percentage]
colors = ['#4CAF50', '#FF5733']


# Create a bar plot of the results
sns.set_style("whitegrid")
sns.set_palette(colors)
ax = sns.barplot(x=labels, y=values)
ax.set_title('Percentage of Movies with Good Ratings that Won an Oscar')
ax.set_xlabel('Average Rating')
ax.set_ylabel('Percentage')
ax.set_ylim([0, 100])

# save the figure
fig = ax.get_figure()
fig.savefig('percentage_movies_good_rating_oscar.png')


plt.show()
```
